```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dateutil import parser
```

```python
df = pd.read_csv("../../../Datas/archivo_limpio.csv", encoding="utf-8")
```

```python
def ventas_diarias_simuladas(nombre):
    nombre = nombre.lower()
    
    # Alta rotación
    if any(x in nombre for x in [
        "agua", "coca", "pepsi", "sprite", "fanta",
        "pan", "tortilla", "leche", "huevo", "azúcar"
    ]):
        return np.random.randint(20, 121)
    
    # Rotación media
    elif any(x in nombre for x in [
        "cereal", "yogurt", "atun", "frijol", "arroz",
        "galleta", "sabritas", "doritos", "ruffles",
        "jugo", "sopa", "chocolate"
    ]):
        return np.random.randint(5, 51)
    
    # Baja rotación
    elif any(x in nombre for x in [
        "detergente", "papel", "aceite", "mayonesa",
        "queso", "helado", "mantequilla"
    ]):
        return np.random.randint(0, 53)
    
    # Premium / alcohol
    elif any(x in nombre for x in [
        "tequila", "mezcal", "vino", "ron", "whisky", "cerveza"
    ]):
        return np.random.randint(0, 30)
    
    # Default razonable
    else:
        return np.random.randint(3, 30)

df["ventas_diarias"] = df["nombre_producto"].apply(ventas_diarias_simuladas)
df.head()
```

```python
columnas_ordenadas = [
    "id_transaccion",
    "fecha",
    "mes",
    "id_producto",
    "nombre_producto",
    "id_tienda",
    "nombre_tienda",
    "categoria_tienda",
    "ciudad_normalizada",
    "ventas_diarias",
    "precio_normalizado"
]

df = df[columnas_ordenadas]
df.head()
df.to_csv('Data_ventas_diarias.csv', index = False)
```
