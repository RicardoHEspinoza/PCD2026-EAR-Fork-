## Lo siguiente que se obtuvo es el catalogo de precios por producto

```python
df2 = pd.read_csv("../../../Datas/4. ciudades_normalizados.csv")
```

```python

reporte_global = df2.groupby(['nombre_producto', 'ciudad_normalizada', 'nombre_tienda'])['precio_normalizado'].agg(['mean', 'min', 'max'])

# Renombramos las columnas para que el CSV sea fácil de entender para otros
reporte_global.columns = ['precio_promedio', 'precio_minimo', 'precio_maximo']

# Ordenamos alfabéticamente por producto y ciudad
reporte_global = reporte_global.sort_index()

# Mostramos una vista previa de los primeros 10 registros
reporte_global.head(10)
```

En este catalogo se obtiene el nombre del producto, la ciudad y el nombre de la tienda ys e le aplican fucniones para obtener el precio por ciudad y por cada tienda de esa ciudad asi como el minimo y el maximo

## Lo siguiente es el obetener los meses gracias a la fecha

```python
df2['fecha'] = pd.to_datetime(df2['fecha'])


df2['mes'] = df2['fecha'].dt.month 
```

```python
df2
```

## Ya que tenemos eso,se calculo el costo promedio de cada producto sin importar su mes, es decir cuanto cuesta en promedio al año

```python
# Calculamos el promedio anual de cada producto
promedio_anual_prod = df2.groupby('nombre_producto')['precio_normalizado'].mean().rename('promedio_general_producto')

# Unimos este promedio al DataFrame original para tenerlo como referencia en cada fila
df2 = df2.merge(promedio_anual_prod, on='nombre_producto')
df2
```
