### Carga del csv

```python
with open ("../Datas/0. ventas_retail_1M.csv", encoding = "utf-8") as file:
    df = pd.read_csv(file)
```

```python
def normalizar_id(col: pd.Series) -> pd.Series:
    return (
        col.astype(str)
           .str.upper()
           .str.strip()
           .str.extract(r"(\d+)", expand=False)
           .astype("Int64")
    )

# Aplicación IN-PLACE
df["id_tienda"]   = normalizar_id(df["id_tienda"])
df["id_producto"] = normalizar_id(df["id_producto"])
```

```python
catalogo_tienda = (
    df.groupby(["id_tienda", "nombre_tienda"])
      .size()
      .reset_index(name="frecuencia")
      .sort_values("frecuencia", ascending=False)
      .drop_duplicates("id_tienda")
      .set_index("id_tienda")["nombre_tienda"]
)

catalogo_producto = (
    df.groupby(["id_producto", "nombre_producto"])
      .size()
      .reset_index(name="frecuencia")
      .sort_values("frecuencia", ascending=False)
      .drop_duplicates("id_producto")
      .set_index("id_producto")["nombre_producto"]
)
```

```python
catalogo_tienda_df = catalogo_tienda.rename("nombre_tienda").reset_index()
catalogo_producto_df = catalogo_producto.rename("nombre_producto").reset_index()
```

```python
df.to_csv(
    "../Datas/1. Data_ID_normalizado.csv",
    index=False,
    encoding="utf-8"
)

catalogo_producto_export = (
    catalogo_producto
        .rename("producto")
        .reset_index()
)

catalogo_producto_export.to_csv(
    "../Catalogos/catalogo_productos.csv",
    index=False,
    encoding="utf-8"
)

catalogo_tienda_export = (
    catalogo_tienda
        .rename("tienda")
        .reset_index()
)

catalogo_tienda_export.to_csv(
    "../Catalogos/catalogo_tiendas.csv",
    index=False,
    encoding="utf-8"
)
```
