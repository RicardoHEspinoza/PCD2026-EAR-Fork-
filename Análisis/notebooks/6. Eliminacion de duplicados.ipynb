```python
# 1. Columnas a comparar
cols = df.columns.difference(['id_transaccion'])

# 2. Máscara para ver TODOS los involucrados en una duplicidad
# keep=False marca todas las apariciones, no solo la segunda
mascara_revision = df.duplicated(subset=cols, keep=False)

# 3. Mostrar el resultado ordenado para comparar uno junto al otro
revision_duplicados = df[mascara_revision].sort_values(by=list(cols))

print(revision_duplicados.head(10))
```

```python
columnas_relevantes = df.columns.difference(['id_transaccion'])


mascara_no_duplicados = ~df.duplicated(subset=columnas_relevantes, keep='first')

df_final = df[mascara_no_duplicados]
```
