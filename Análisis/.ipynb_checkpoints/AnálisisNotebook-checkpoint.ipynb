Primeramente debemos rellenar los datos faltantes, en el caso de las ciudades vamos a rellenar de acuerdo al id de la tienda, para ello primero debemos sacar que id le corresponde a cada tienda, de paso contamos igual cuantos productos hay. Entonces vamos a generar un data de que tiendas hay y los productos que hay

```python
tiendas_unicas = df['id_tienda'].nunique()
productos_unicos = df['id_producto'].nunique()

print(f"Número de tiendas únicas: {tiendas_unicas}")
print(f"Número de productos únicos: {productos_unicos}")
```

esto conto unicamente los no repetidos, pero al haber un pequeno cambio en la notacion el conteo puede cambiar, asi que primero debemos normalizar los datos de los id de producto y de las tiendas para tomarlo como base para hacer la agrupacion.

```python
df['id_producto'].isna().sum()
df['id_tienda'].isna().sum()

df['id_producto'].astype(str).str.contains(r'^\s|\s$', regex=True).sum()
df['id_tienda'].astype(str).str.contains(r'^\s|\s$', regex=True).sum()

df['id_producto'].astype(str).str.contains(r'\D').sum()
df['id_tienda'].astype(str).str.contains(r'\D').sum()

df['id_producto'].astype(str).str.match(r'^0+\d+').sum()
df['id_tienda'].astype(str).str.match(r'^0+\d+').sum()

id_prod_strip = df['id_producto'].astype(str).str.strip()
id_prod_strip.nunique(), df['id_producto'].nunique()


(
    df.groupby('id_producto')['nombre_producto']
      .nunique()
      .reset_index(name='nombres_distintos')
      .query('nombres_distintos > 1')
)
(
    df.groupby('id_tienda')['nombre_tienda']
      .nunique()
      .reset_index(name='nombres_distintos')
      .query('nombres_distintos > 1')
)
```

```python
resumen_errores = pd.DataFrame({
    "Error": [
        "Nulos",
        "Espacios",
        "No numéricos",
        "Ceros a la izquierda",
        "Formato inconsistente",
        "ID con múltiples entidades"
    ],
    "Producto": [
        df['id_producto'].isna().sum(),
        df['id_producto'].astype(str).str.contains(r'^\s|\s$', regex=True).sum(),
        df['id_producto'].astype(str).str.contains(r'\D').sum(),
        df['id_producto'].astype(str).str.match(r'^0+\d+').sum(),
        df['id_producto'].nunique() - id_prod_strip.nunique(),
        df.groupby('id_producto')['nombre_producto'].nunique().gt(1).sum()
    ],
    "Tienda": [
        df['id_tienda'].isna().sum(),
        df['id_tienda'].astype(str).str.contains(r'^\s|\s$', regex=True).sum(),
        df['id_tienda'].astype(str).str.contains(r'\D').sum(),
        df['id_tienda'].astype(str).str.match(r'^0+\d+').sum(),
        df['id_tienda'].nunique() - df['id_tienda'].astype(str).str.strip().nunique(),
        df.groupby('id_tienda')['nombre_tienda'].nunique().gt(1).sum()
    ]
})

resumen_errores
```

Efectivamente los ID tanto de los productos como de las tiendas tienen errores, entonces tenemos que corregirlos:
- Los espacios se pueden resolver con strip()
- Los no numericos abria que ver de que tipos son y si se pueden corregir
- Los ceros a la izquierda tambien son corregibles
- Los formatos inconsistentes abria que ver de que tipos son
- para los ID multiples abria que hacer un conteo de productos y ver que ID relacionado con producto y tienda se repiten mas y corregir los incorrectos

```python
#Corregir los espacios en los IDs
df['id_producto_raw'] = df['id_producto']
df['id_tienda_raw'] = df['id_tienda']

df['id_tienda_clean'] = (
    df['id_tienda']
    .astype(str)
    .str.strip()
)
#detectamos cuales no tienen problemas
mask_prod_num = df['id_producto_clean'].str.fullmatch(r'\d+')
mask_tienda_num = df['id_tienda_clean'].str.fullmatch(r'\d+')

df.loc[mask_prod_num, 'id_producto_clean'] = (
    df.loc[mask_prod_num, 'id_producto_clean']
      .str.lstrip('0')
      .replace('', '0')
)

df.loc[mask_tienda_num, 'id_tienda_clean'] = (
    df.loc[mask_tienda_num, 'id_tienda_clean']
      .str.lstrip('0')
      .replace('', '0')
)
```

```python
#Creamos las columnas normalizadas
df['id_producto_norm'] = df['id_producto_clean']
df['id_tienda_norm'] = df['id_tienda_clean']


print("RESUMEN PRODUCTOS")
print(f"Productos antes:    {df['id_producto_raw'].nunique()}")
print(f"Productos después:  {df['id_producto_norm'].nunique()}")
print(f"Reducción:          {df['id_producto_raw'].nunique() - df['id_producto_norm'].nunique()}")

print("\nRESUMEN TIENDAS")
print(f"Tiendas antes:      {df['id_tienda_raw'].nunique()}")
print(f"Tiendas después:    {df['id_tienda_norm'].nunique()}")
print(f"Reducción:          {df['id_tienda_raw'].nunique() - df['id_tienda_norm'].nunique()}")


```

El siguiente error a arreglar debe ser el de multiples identidades, para ello vamos a sacar una meustra de frecuencia, es decir ver para que ID tanto de producto como tienda que tienda u objeto se repite mas, y sacado eso vamos a reemplazar los ID erroneos por ese ID

```python
conteo_tiendas = (
    df.groupby(['id_tienda_norm', 'nombre_tienda'])
      .size()
      .reset_index(name='conteo')
)

conteo_productos = (
    df.groupby(['id_producto_norm', 'nombre_producto'])
      .size()
      .reset_index(name='conteo')
)
```

```python
total_tiendas = (
    conteo_tiendas
    .groupby('id_tienda_norm')['conteo']
    .sum()
    .reset_index(name='total')
)
total_productos = (
    conteo_productos
    .groupby('id_producto_norm')['conteo']
    .sum()
    .reset_index(name='total')
)
```

```python
tienda_dom = (
    conteo_tiendas
    .sort_values(['id_tienda_norm', 'conteo'], ascending=[True, False])
    .groupby('id_tienda_norm')
    .first()
    .reset_index()
)

producto_dom = (
    conteo_productos
    .sort_values(['id_producto_norm', 'conteo'], ascending=[True, False])
    .groupby('id_producto_norm')
    .first()
    .reset_index()
)
```

```python
tienda_dom = tienda_dom.merge(total_tiendas, on='id_tienda_norm')
tienda_dom['porcentaje'] = tienda_dom['conteo'] / tienda_dom['total']

producto_dom = producto_dom.merge(total_productos, on='id_producto_norm')
producto_dom['porcentaje'] = producto_dom['conteo'] / producto_dom['total']
```

```python
tiendas_confiables = tienda_dom[tienda_dom['porcentaje'] >= 0.9]
productos_confiables = producto_dom[producto_dom['porcentaje'] >= 0.9]

tiendas_conflictivas = tienda_dom[tienda_dom['porcentaje'] < 0.9]
productos_conflictivos = producto_dom[producto_dom['porcentaje'] < 0.9]
```

```python
map_tienda = (
    tiendas_confiables
    .set_index('id_tienda_norm')['nombre_tienda']
    .to_dict()
)
map_producto = (
    productos_confiables
    .set_index('id_producto_norm')['nombre_producto']
    .to_dict()
)
```

```python
df['nombre_tienda_corr'] = df['id_tienda_norm'].map(map_tienda)
df['nombre_producto_corr'] = df['id_producto_norm'].map(map_producto)
```

```python
df['nombre_tienda_final'] = df['nombre_tienda_corr'].fillna(df['nombre_tienda'])
df['nombre_producto_final'] = df['nombre_producto_corr'].fillna(df['nombre_producto'])
```

```python
impacto_tienda = (
    df['nombre_tienda'] != df['nombre_tienda_final']
).sum()
impacto_producto = (
    df['nombre_producto'] != df['nombre_producto_final']
).sum()
```

En teoria ya hemos normalizados los ID, pero ahora queremos saber que ID corresponde a cada cosa, para ello vamos a crear dos csv con los ID de las tiendas y con los ID de los productos

```python
tabla_tiendas = (
    df[['id_tienda_norm', 'nombre_tienda_final']]
    .drop_duplicates()
    .sort_values(by='id_tienda_norm')
    .rename(columns={
        'id_tienda_norm': 'id_tienda',
        'nombre_tienda_final': 'tienda'
    })
)
tabla_productos = (
    df[['id_producto_norm', 'nombre_producto_final']]
    .drop_duplicates()
    .sort_values(by='id_producto_norm')
    .rename(columns={
        'id_producto_norm': 'id_producto',
        'nombre_producto_final': 'producto'
    })
)


```

```python
tabla_tiendas.to_csv("dim_tiendas.csv", index=False)
tabla_productos.to_csv("dim_productos.csv", index=False)

print(tabla_tiendas.head())
print(tabla_tiendas.tail())
print(tabla_productos.head())
print(tabla_productos.tail())
```

Esto nos dice que si bien los ID ya estan normalizados existen otro tipo de ID que tienen explicito escrito ID. Entonces tenemos que reemplazarlos tambien

```python
mask_prod_prefijo = df['id_producto_raw'].str.match(r'(?i)^ID\d+$', na=False)

df.loc[mask_prod_prefijo, ['id_producto_raw', 'nombre_producto']].head(10)

mask_tienda_prefijo = df['id_tienda_raw'].str.match(r'(?i)^ID\d+$', na=False)

df.loc[mask_tienda_prefijo, ['id_tienda_raw', 'nombre_tienda']].head(10)
```

```python
df.loc[mask_prod_prefijo, 'id_producto_norm'] = (
    df.loc[mask_prod_prefijo, 'id_producto_raw']
      .str.upper()
      .str.replace('ID', '', regex=False)
)

df.loc[mask_tienda_prefijo, 'id_tienda_norm'] = (
    df.loc[mask_tienda_prefijo, 'id_tienda_raw']
      .str.upper()
      .str.replace('ID', '', regex=False)
)

```

```python
tabla_tiendas = (
    df[['id_tienda_norm', 'nombre_tienda_final']]
    .drop_duplicates()
    .sort_values(
        by='id_tienda_norm',
        key=lambda x: x.astype(int)  # orden numérico real
    )
    .rename(columns={
        'id_tienda_norm': 'id_tienda',
        'nombre_tienda_final': 'tienda'
    })
)
tabla_productos = (
    df[['id_producto_norm', 'nombre_producto_final']]
    .drop_duplicates()
    .sort_values(
        by='id_producto_norm',
        key=lambda x: x.astype(int)  # orden numérico real
    )
    .rename(columns={
        'id_producto_norm': 'id_producto',
        'nombre_producto_final': 'producto'
    })
)

tabla_tiendas.to_csv("dim_tiendas.csv", index=False)
tabla_productos.to_csv("dim_productos.csv", index=False)

print(tabla_tiendas.head())
print(tabla_tiendas.tail())
print(tabla_productos.head())
print(tabla_productos.tail())
```

Esto nos dice que ya todos los ID estan normalizados, asi que volvemos a aplicar un analisis para ver si ya no quedan problemas

```python
def contar_errores(col):
    return {
        "Nulos": col.isna().sum(),
        "Espacios": col.astype(str).str.match(r'^\s+|\s+$', na=False).sum(),
        "No numéricos": (~col.astype(str).str.fullmatch(r'\d+', na=False)).sum(),
        "Ceros a la izquierda": col.astype(str).str.match(r'^0+\d+', na=False).sum(),
        "Formato inconsistente": col.astype(str).str.match(r'(?i)^ID\d+', na=False).sum(),
    }

errores_antes = pd.DataFrame({
    "Producto": contar_errores(df['id_producto_raw']),
    "Tienda": contar_errores(df['id_tienda_raw'])
})
errores_antes

```

```python
errores_despues = pd.DataFrame({
    "Producto": contar_errores(df['id_producto_norm']),
    "Tienda": contar_errores(df['id_tienda_norm'])
})
print (errores_despues)


```
